```python
def gerar_tabela(df, coluna, emprestimos):
  tabela = df[[emprestimos, coluna]].groupby(coluna).count().reset_index().sort_values(by=emprestimos, ascending=False)
  total = tabela[emprestimos].sum()
  tabela['porcentagem'] = tabela[emprestimos].apply(lambda x : x/ total)
  tabela['porcentagem'] = tabela['porcentagem'].apply(lambda x : x * 100)
  tabela['porcentagem'] = tabela['porcentagem'].apply(round)
  tabela.rename(columns={'id_exemplar' : 'numero_emprestimos'}, inplace=True)
  tabela = tabela.reset_index(drop=True)
  return tabela
```

```python
# Pergunta: “Como se distribuem os empréstimos de exemplares pelos tipos de vínculo dos usuários?”

gerar_tabela(df_final,'tipo_vinculo_usuario','id_exemplar')
```

```python
# Quais coleções são mais emprestadas?

gerar_tabela(df_final,'colecao','id_exemplar')
```

```python
# Quais são as bibliotecas com mais ou menos quantidade de empréstimos?

gerar_tabela(df_final,'biblioteca','id_exemplar')
```

```python
# De quais temas da CDU são os exemplares emprestados?

gerar_tabela(df_final,'CDU','id_exemplar')
```

```python
df_previsao = pd.read_table('https://raw.githubusercontent.com/FranciscoFoz/7_Days_of_Code_Alura-Python-Pandas/main/Dia_7-Apresentando_resultados_em_HTML/Dataset/previsao')
df_previsao = df_previsao['curso previsao_2022'].str.split(expand=True)
df_previsao.rename(columns={0: 'Curso', 1: 2022}, inplace=True)
df_previsao['Curso'].replace('_', ' ', regex=True, inplace=True)
df_previsao['Curso'] = df_previsao['Curso'].str.title()
```

```python
df_pos_2017_2022 = df_pos_2.merge(df_previsao, left_index=True, right_index=True, suffixes=('', '_y'))
df_pos_2017_2022 = df_pos_2017_2022[['Curso', 2017, 2018, 2019, 2022]]
df_pos_2017_2022[2022] = df_pos_2017_2022[2022].astype(int)
```

```python
def variacao_porcentagem(coluna_1, coluna_2,df=df_pos_2017_2022):
  x = ((df[coluna_2] * 100) / df[coluna_1])
  x = x - 100
  return x.round(decimals=2)
```

```python
df_pos_2017_2022['2017~2018'] = variacao_porcentagem(2017, 2018)
df_pos_2017_2022['2018~2019'] = variacao_porcentagem(2018, 2019)
df_pos_2017_2022['2019~2022'] = variacao_porcentagem(2019, 2022)
df_pos_2017_2022 = df_pos_2017_2022[['Curso','2017~2018', '2018~2019', '2019~2022']]
df_pos_2017_2022 = df_pos_2017_2022.set_index('Curso')
```
